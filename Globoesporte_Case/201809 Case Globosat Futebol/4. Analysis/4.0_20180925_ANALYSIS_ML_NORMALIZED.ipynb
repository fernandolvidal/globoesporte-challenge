```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix, classification_report
import eli5
from eli5.sklearn import PermutationImportance

np.random.seed(42)

dir_path = '../2. Prepared Data/'
save_path = dir_path
file_name = ['20180924_merged_train_sampled_simple.csv','20180924_merged_test_sampled_simple.csv']
```

## <u>Modelagem</u>:

```python
#SEPARANDO EM TREINAMENTO E TESTE
train_sample = pd.read_csv(dir_path+file_name[0])
test_sample = pd.read_csv(dir_path+file_name[1])

train_X = train_sample.drop(['TIME','KEY'],axis=1)
train_y = train_sample.TIME
test_X = test_sample.drop(['TIME','KEY'],axis=1)
test_y = test_sample.TIME

features = list(train_X.columns)
```

```python
#TESTE SIMPLES COM SVC
scaler = MinMaxScaler().fit(train_X)
normalized_train_X = scaler.transform(train_X)
normalized_test_X = scaler.transform(test_X)
```

```python
svc = LinearSVC().fit(train_X,train_y)
```

```python
pred = svc.predict(test_X)
ac =np.mean(pred == test_y)
```

```python
print(classification_report(test_y,pred));
cm = pd.DataFrame(confusion_matrix(test_y,pred),columns= list(np.unique(test_y)),index=list(np.unique(test_y)))
cm.head(30)
```

```python
#TESTE COM DADOS NORMALIZADOS
svc_norm = LinearSVC().fit(normalized_train_X,train_y)
```

```python
pred = svc_norm.predict(normalized_test_X)
ac =np.mean(pred == test_y)
```
