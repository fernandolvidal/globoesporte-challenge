## <u> Leitura dos dados</u>:

```python
import numpy as np
import pandas as pd

#CAMINHOS E DIRETORIOS PADROES
dir_path = '../1. Original Data'
save_path = '../2. Prepared Data'
train_pattern = '/TRAIN_BD_'
test_pattern = '/TEST_BD_'
file_name = ['CONSUMO.csv','DEM_TIME.csv']

#LEITURA DOS DADOS ORIGINAIS - SEM SEPARACAO ENTRE TREINO E TESTE
consumo = pd.read_csv(dir_path+train_pattern+file_name[0]).append\
         (pd.read_csv(dir_path+test_pattern+file_name[0]))
usuario = pd.read_csv(dir_path+train_pattern+file_name[1]).append\
         (pd.read_csv(dir_path+test_pattern+file_name[1]))
```

## <u> Padronizacao Base de Usuarios</u>:

```python
#FUNCOES
def standardize_team(team):
    if team.lower() == 'goiais': return 'goias'
    return team.lower().replace(' ','-')

#SELECAO DE MELHORES ATRIBUTOS
usuario_std = usuario.iloc[:,[0,-1]].copy()

#PADRONIZACAO
usuario_std.iloc[:,-1] = usuario.TIME.apply(standardize_team).values
usuario_std.set_index('KEY',inplace=True)

#POSSIVEIS CLASSES
times = list(np.unique(usuario_std.TIME))
usuario_std.head()
```

## <u> Engenharia de Features sob Base de Consumo</u>:

```python
#FUNCOES

#FUNCAO SIMPLES DE EXTRACAO DE TIME DO URL
def get_team_in_dashes(url):
    for time in times:
        if 'times/'+time+'/' in str(url): 
            return time
           
#FUNCAO DE PROCESSAMENTO DE URL
def proc_url(data,f):
    return data.apply(f)

#PROCESSAMENTO 

#APLICANDO PROCESSAMENTO SIMPLES DE EXTRACAO DE URL
consumo_proc = consumo.copy()
consumo_proc['TIME'] = proc_url(consumo.url,get_team_in_dashes)   

#RETIRANDO URLS IRRELEVANTES
consumo_proc = consumo_proc.dropna()

#APLICANDO PROCESSAMENTO DE FORMATO DA TABELA
consumo_proc = consumo_proc.groupby(by=('KEY','TIME')).sum()
consumo_proc = consumo_proc.reset_index()
consumo_proc = consumo_proc.pivot(index='KEY',columns='TIME',values='tempo')
consumo_proc = consumo_proc.fillna(0)

consumo_proc.head(100)
```

## <u> Uniao de bases</u>:

```python
#ESTRUTURA (TEMPO GASTO NA PAGINA DO TIME), (TIME DO USUARIO)
merged = consumo_proc.join(usuario_std,how='left')
merged.head(100)
```

## <u> Modelagem Intuitiva</u>:

```python
#GERANDO MODELO BASEADO NO TIME EM QUE O USUARIO MAIS PERDE TEMPO NA PAGINA
data_for_pred = merged.drop('TIME',axis=1)
data_for_pred['pred'] = data_for_pred.idxmax(axis=1)
data_for_pred['y'] = merged.TIME

acc = np.mean(data_for_pred.pred == data_for_pred.y)
print(acc)
```
